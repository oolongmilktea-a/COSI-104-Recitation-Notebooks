### COSI 104 Seaborn Visualization Using Iris Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# iris dataset
iris = sns.load_dataset("iris")

sns.set_theme(context='notebook', style='darkgrid', palette='pastel', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

#### Histogram
- to find the distribution of a single numerical feature. We can learn the range of values, which features are separable, and if there is a presence of a skew or outlier

```python
iris.hist(figsize=(10, 8), bins=20)
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.show()
```

#### Scatterplot
- use it to find a relationship between two numerical variables. We can find patterns, trends, and potential clustering. 
- could find out whether there is a linear or non linear relationship, how well the classes are separable in two dimensions, and if there are outliers or any unusual points

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", style="species", size="petal_length", sizes=(20, 200))
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", style="species", s=150, alpha=0.6)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

#### Pair Plot
- we are shown a scatter plot for every pair of features, with a histogram/KDE on the diagonals. 
- we can learn which features mattr most, whether dimensionality reduction makes sence or feature selection too. Pretty good choice for a plot for a quick data exploratoin

```python
sns.pairplot(iris, hue="species")
plt.show()
```

#### Swarm Plot
- Shows every individual data point grouped by categorical variables. We can find exact data distribution, small sample behaviours, overlaps, and crowding. 
- Quite good fro smaller datasets as it can get quite messy for larger ones.

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(
    data=iris,
    x="species",
    y="petal_length"
)
plt.title("Swarm Plot of Petal Length by Species")
plt.show()
```

#### Box Plot
- your standard high school box plot with median, iqrs, and outliers. Good to find behaviours at the center, the spreads, and see outliers and skewness.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=iris,
    x="species",
    y="petal_width"
)
plt.title("Box Plot of Petal Width by Species")
plt.show()
```

#### Violin Plot
- distribution shape and summary statistics, it basically combines a box plot and KDE (continuous histogram, Kernel Density Estimate).
- can see the density of values at each level, whether the data is symmetric or skews, differences in distribution shape between groups. 
- narrow means tighter clustering and wider show more variance

```python
plt.figure(figsize=(8, 5))
sns.violinplot(
    data=iris,
    x="species",
    y="sepal_length"
)
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

#### Joint Plot
- scatter plot + marginal distributions.
- see univariate and bivariate insights, see the densities vs point concentrations and the strength of the correlations. Can find separability too

```python
sns.jointplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    kind="scatter"
)
plt.show()
```

```python
sns.jointplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    kind="hex", ratio=2, marginal_ticks=True, height=5
)
plt.show()
```

#### KDE Plot
- Smooth probability density estimations, basically a continuous version of a histogram. We can see the distribution shapes without binning artifacts, where values concentrate, and the class overlaps.

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(
    data=iris,
    x="petal_length",
    hue="species",
    fill=True
)
plt.title("KDE Plot of Petal Length")
plt.show()
```

#### Heatmap (Correlation Matrix)
- Pairwaise correlations between the numerical features, and see the strength and direction of the linear relationships.
- see redundant deatures, which features move together, and determine whether linear models would work well for the dataset
- probably my most used for EDA, I always generate one

```python
plt.figure(figsize=(6, 5))
corr = iris.drop(columns="species").corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    square=True
)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

#### Strip plot
- relationships between the categorical and continuous variables, and the summary of the continuous variable

```python
sns.stripplot(data=iris, x='petal_length', hue='species', y='sepal_length', s=25, alpha=0.2,
             jitter=False, marker='D',palette='viridis')
plt.title('StripPlot')
plt.show()
```

#### Takeaways for Iris
- petal length and petal width are the more informative features for classifying the Iris, as they show a lot of correlation as well as a strong separation. Setosa classification is linearly separable from the other two species whereas the other two, Versicolor and Virginica, show some overlaps. Sepal measurements have less say in the overall classifications and have a lot of overlap for each of the species, which makes it a less relevant feature in our case.
